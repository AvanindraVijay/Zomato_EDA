```python
country_name = Merge_df.Country.value_counts().index
```

```python
country_val = Merge_df.Country.value_counts().values
```

```python
#Pie Chart top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_name[:3])
```

```python
Merge_df.columns
```

```python
Merge_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings = Merge_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings.head()
```

#### Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating' , y = 'Rating Count', data = Ratings)
```

```python
sns.barplot(x='Aggregate rating', y = 'Rating Count', hue='Rating color', data = Ratings,palette=['blue','red','orange','yellow','green','green']) 
```

### Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
#Find the top 10 cuisines
Merge_df.Cuisines.value_counts().head(10)
```

```python
cuisine_counts = Merge_df.groupby(['City', 'Cuisines']).size().reset_index(name='Count')
```

```python
cuisine_counts.head(10)
```

```python
cuisine_values = Merge_df.Cuisines.value_counts().values[:5]
cuisine_labels = Merge_df.Cuisines.value_counts().index[:5] 
plt.pie(x=cuisine_values, labels = cuisine_labels,autopct = '%1.2f%%')
```

```python
plt.rcParams['figure.figsize'] = (20, 5)
sns.barplot(x='cuisine_values', y='cuisine_counts', hue='City', data=cuisine_counts)
plt.show()
```
